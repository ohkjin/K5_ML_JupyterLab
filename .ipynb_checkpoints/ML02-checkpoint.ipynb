```python
#fig: 바깥 도화지
fig, axes = plt.subplots(1,3, figsize=(10,3))         # subplots(row,col,figsize=(도화지 크기 w,h))

for n_neighbors, ax in zip([1,3,9],axes):             # in zip(n_neighbors array, ax array) 
    print(n_neighbors)
    print(ax)
    #fit method returns self object
    clf = KNeighborsClassifier(n_neighbors=n_neighbors).fit(X,y)
    mglearn.plots.plot_2d_separator(clf, X, fill=True, eps=0.5, ax=ax, alpha=.4)  # fill: 배경 다채우기, eps=비율
    mglearn.discrete_scatter(X[:,0],X[:,1],y,ax=ax)                               # 배경에 점찍기
    ax.set_title("{}{} 이웃".format(n_neighbors,n_neighbors))
    ax.set_xlabel("0")
    ax.set_ylabel("1")
axes[0].legend(loc=3)
```

```python
from sklearn.datasets import load_breast_cancer

cancer = load_breast_cancer()
X_train, X_test, y_train, y_test = train_test_split(cancer.data, cancer.target, stratify=cancer.target, random_state=66)

training_accuracy = []
test_accuracy = []

#1 에서 10 까지 n_neighbors
neighbors_settings=range(1,11)

for n_neighbors in neighbors_settings:
    #Create Model
    clf = KNeighborsClassifier(n_neighbors=n_neighbors)
    clf.fit(X_train, y_train)
    #훈련 세트 정확도 저장
    training_accuracy.append(clf.score(X_train, y_train))
    #일반화 정확도 저장
    test_accuracy.append(clf.score(X_test, y_test))

plt.plot(neighbors_settings, training_accuracy,label="훈련정확도")
plt.plot(neighbors_settings, test_accuracy, label="Test 정확도")
plt.ylabel("Accuracy")
plt.xlabel("n_neighbors")
plt.legend()
```

### KNN 모델 평가

#### 선형회귀(LinearRegression)

##### 단순한 모델

```python
from sklearn.linear_model import LinearRegression
X,y = mglearn.datasets.make_wave(n_samples=60)
X_train, X_test, y_train, y_test = train_test_split(X,y,random_state=42)
lr = LinearRegression().fit(X_train,y_train)
```

```python
print("coef:",lr.coef_)            # coefficient 계수, weight 가중치, 기울기
print("intercept:", lr.intercept_) # intercept   절편, offset 편향
```

```python
print("훈련",lr.score(X_train,y_train))
print("테스트",lr.score(X_test,y_test))
```

##### 복잡한 모델
선형 모델: 복잡한 데이터에는 사용할 수 없다

```python
# 복잡한 모델
X,y = mglearn.datasets.load_extended_boston()
X_train, X_test, y_train, y_test = train_test_split(X,y,random_state=42)
lr = LinearRegression().fit(X_train,y_train)
print("훈련 ",lr.score(X_train,y_train))
print("테스트",lr.score(X_test,y_test))
print("복잡한 모델에는 사용할 수 없다")
```

#### L2 규제 (Ridge Regression)
coef를 0에 가깝게 만들어서 모델을 단순화

```python
from sklearn.linear_model import Ridge
from sklearn.model_selection import train_test_split
X,y = mglearn.datasets.load_extended_boston()
X_train, X_test, y_train, y_test = train_test_split(X,y,random_state=42)
ridge = Ridge(alpha=10).fit(X_train, y_train) # alpha 값이 높으면(default=1.0) coef를 0에 가깝게 만듬 -> 훈련 성능 대비 모델 단순화
```

```python
print("훈련",ridge.score(X_train,y_train))
print("테스트",ridge.score(X_test,y_test))
```

#### L1 규제 (Lasso Regression)
coef를 0으로 만들어서 모델을 단순화

```python
from sklearn.linear_model import Lasso
from sklearn.model_selection import train_test_split
X,y = mglearn.datasets.load_extended_boston()
X_train, X_test, y_train, y_test = train_test_split(X,y,random_state=42)
lasso = Lasso(alpha=0.1).fit(X_train, y_train) 
```

```python
print("훈련",lasso.score(X_train,y_train))
print("테스트",lasso.score(X_test,y_test))
print("사용 특성 수",np.sum(lasso.coef_!=0))
```
